## PREGUNTA 7

7.- Implemente el algoritmo de Grover para determinar la máxima cantidad de aristas en un grafo de 5 nodos y 6 aristas:

\begin{align}
(0,3)\\
(0,4)\\
(1,3)\\
(1,4)\\
(2,3)\\
(2,4)\\
\end{align}

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit.circuit.library import XGate,ZGate
from qiskit.visualization import plot_histogram
from qiskit import execute, Aer
```

```python
def oracle_part(qc,qr):
    mcx = XGate().control(3)
    
    #Los qubits del 0 al 4 son los nodos.

    #Los qubits del 5 al 10 nos permiten reconocer si la arista es correcta
    #(i.e. si se da entre nodos de distinto color).
    qc.cx(qr[0], qr[5])
    qc.cx(qr[3], qr[5])
    qc.barrier()
    qc.cx(qr[0], qr[6])
    qc.cx(qr[4], qr[6])
    qc.barrier()
    qc.cx(qr[1], qr[7])
    qc.cx(qr[3], qr[7])
    qc.barrier()
    qc.cx(qr[1], qr[8])
    qc.cx(qr[4], qr[8])
    qc.barrier()
    qc.cx(qr[2], qr[9])
    qc.cx(qr[3], qr[9])
    qc.barrier()
    qc.cx(qr[2], qr[10])
    qc.cx(qr[4], qr[10])
    qc.barrier()
    

    #SUMADOR (Entrada 6 qubits) (salida 3 qubits)
    #Los qubits del 11 al 13 nos van a permitir contar la cantidad de aristas correctas 
    #de tal forma que la suma será expresada en binario con cifras dadas por q_11, q_10 y q_9.
    #(Con 3 qubits podemos sumar hasta 7 en binario, lo cual es suficiente ya que hay maximo 6 aristas)
    qc.barrier()    
    qc.cx(qr[5],qr[11])
    qc.barrier()
    
    qc.ccx(qr[6], qr[11], qr[12])
    qc.cx(qr[6],qr[11])
    qc.barrier()
    
    qc.ccx(qr[7], qr[11], qr[12])
    qc.cx(qr[7],qr[11])
    qc.barrier()
    
    qc.append(mcx,[8,11,12,13])
    qc.ccx(qr[8], qr[11], qr[12])
    qc.cx(qr[8],qr[11])
    qc.barrier()
    
    qc.append(mcx,[9,11,12,13])
    qc.ccx(qr[9], qr[11], qr[12])
    qc.cx(qr[9],qr[11])
    qc.barrier()
    
    qc.append(mcx,[10,11,12,13])
    qc.ccx(qr[10], qr[11], qr[12])
    qc.cx(qr[10],qr[11])
    qc.barrier()


    #
    qc.x(qr[11])
    qc.barrier()
    
    #El qubit 14 indicará cuando se haya encontrado un estado deseado
    qc.append(mcx,[11,12,13,14])
    
    return qc
```

Nota: Utilicé el siguiente diagrama, para implementar el sumador con entrada de 6 qubits

<img src=Sumador_6q.jpg>

```python
def oracle(qc,qr):
    
    qc=oracle_part(qc,qr)
    qc1=qc.inverse()
    qc.z(14)
    qc.compose(qc1,inplace=True)
    
    return qc
```

```python
def grover_diffusion(qc,qr):
    mcz = ZGate().control(4)
    
    qc.barrier()
    qc.h(range(5))
    qc.x(range(5))
    qc.append(mcz,list(range(0,5)))
    qc.x(range(5))
    qc.h(range(5))
    qc.barrier()
    return qc
```

```python
def grover(qc,qr):
    return grover_diffusion(oracle(qc,qr),qr)
```

```python
qr7=QuantumRegister(15)
qc7=QuantumCircuit(qr7)

qra=QuantumRegister(15)
cra=ClassicalRegister(5)
qa=QuantumCircuit(qra,cra)

qa.h(range(5))
qb=grover(qc7,qr7)

#k=2  (aplico oraculo y difusor 2 veces)
qa.compose(qb.compose(qb),inplace=True)

qa.measure(range(5),cra)

qa.draw()
```

```python
qasm_simulator = Aer.get_backend('qasm_simulator')

job = execute(qa, qasm_simulator, shots = 1000)
counts = job.result().get_counts(qa)

plot_histogram(counts)
```

Por tanto, el algoritmo de Grover entrega con mayor probabilidad los estados 11100 y 00011

Es decir, se maximiza las aristas cuando los nodos (0,1,2) formen una región, y los nodos (3,4) otra región.

## PREGUNTA 9

La implementación de QFT se probó preparando el estado de Fourier $|\widetilde{5}\rangle$ para el cual $\text{QFT}^{\dagger}|\widetilde{5}\rangle= |101\rangle$. Intenta encontrar el estado $|a\rangle$ tal que $\text{QFT}^{\dagger}|a\rangle = |100\rangle$.


Desarrollo: 

Una forma rápida de obtener esto es sabiendo que QFT por ser unitario, se cumple que:  $\text{QFT} .\text{QFT}^{\dagger}=\text{I}$

Así obtenemos que $|a\rangle =\text{QFT}|100\rangle$.

Pero bien, seguiremos la misma línea utilizada en clase para encontrar $|a\rangle$

Traemos las funciones definidas en el notebook de la clase para poder verificar:

```python
def qft_rotations(circuit, n):
    """Performs qft on the first n qubits in circuit (without swaps)"""
    if n == 0:
        return circuit
    n -= 1
    circuit.h(n)
    for qubit in range(n):
        circuit.cp(pi/2**(n-qubit), qubit, n)
    qft_rotations(circuit, n)

#QFT general 
def swap_registers(circuit, n):
    for qubit in range(n//2):
        circuit.swap(qubit, n-qubit-1)
    return circuit

def qft(circuit, n):
    """QFT on the first n qubits in circuit"""
    qft_rotations(circuit, n)
    swap_registers(circuit, n)
    return circuit

#QFT Inversa
def inverse_qft(circuit, n):
    """Does the inverse QFT on the first n qubits in circuit"""
    # First we create a QFT circuit of the correct size:
    qft_circ = qft(QuantumCircuit(n), n)
    # Then we take the inverse of this circuit
    invqft_circ = qft_circ.inverse()
    # And add it to the first n qubits in our existing circuit
    circuit.append(invqft_circ, circuit.qubits[:n])
    return circuit.decompose() # .decompose() allows us to see the individual gates
```

Encontraremos estado $|a\rangle$ tal que $\text{QFT}^{\dagger}|a\rangle = |100\rangle$.

```python
nqubits=3
number=4 #Ya que 4 es 100 en binario
qc9= QuantumCircuit(nqubits)

#llevamos base de Fourier, donde los estados son superposiciones de diferentes frecuencias de fase
for qubit in range(nqubits):
    qc9.h(qubit)

qc9.p(number*pi/4,0)
qc9.p(number*pi/2,1)
qc9.p(number*pi,2)

qc9.draw(output='mpl')
```

Veremos el estado $|a\rangle$ en la esfera de Bloch:

```python
qc_init = qc9.copy()
qc_init.save_statevector()
sim = Aer.get_backend("aer_simulator")
statevector = sim.run(qc_init).result().get_statevector()
plot_bloch_multivector(statevector)
```

Obtenemos el vector de estado pedido en notación braket.
Notar que $|a\rangle$ vive en la base de Fourier, por lo que estará en superposición, y solo nos interesa las fases

```python
backend = Aer.get_backend('statevector_simulator')
job = backend.run(qc9)
result = job.result()
outputstate = result.get_statevector(qc9, decimals=3)
ket_fourier = Statevector(outputstate)
ket_fourier.draw('Latex')
```

Ahora verificamos, que al aplicar $\text{QFT}^{\dagger}|a\rangle$ y medir nos debe dar $|100\rangle$.

```python
qc9_1 = inverse_qft(qc9, nqubits)
qc9_1.measure_all()
qc9_1.draw(output='mpl')
```

```python
results = sim.run(qc9_1).result()
answer = results.get_counts()
plot_histogram(answer)
```
